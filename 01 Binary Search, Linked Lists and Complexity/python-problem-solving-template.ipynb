### 8. Implement the solution and test it using example inputs. Fix bugs, if any.

```python
def binary_search(lo, hi, condition):
  while lo <= hi:
    mid = (lo + hi) // 2
    result = condition(mid)
    # print('lo: ', lo, 'hi: ', hi, 'mid: ', mid)
    # print('result:', result)
    if result == 'found':
      return mid
    elif result == 'move_left':
      hi = mid - 1
    elif result == 'move_right':
      lo = mid + 1
  return -1


def index_of_target(nums, target):
  
  def condition(mid):
    if nums[mid] == target:
      if mid > 0 and nums[mid-1] == target:
        return 'move_left'
      else:
        return 'found'
    elif nums[mid] < target:
      return 'move_right'
    return 'move_left'

  return binary_search(0, len(nums)-1, condition)
```

### 9. Analyze the algorithm's complexity and identify inefficiencies, if any.

1. Doesn't consider all target values in an array in case if there are multiple of them, so it only gives first position and not the last onem where it ends. However, it can be modified to this
2. Complexity is **`O(log(N))`**. Since:
first iteration: `N/2`
second iteration: `N/2^2`
and so on... until we get final index of size `1`, so
`N/2^k = 1`, it means `k = log(N)` complexity.
In terms of memory complexity it's **`O(1)`**
